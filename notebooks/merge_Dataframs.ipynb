```python
import pandas as pd
import os
DATA_PATH = "/home/prsood/projects/def-whkchun/prsood/multi-modal-emotion/data/"
```

```python
df = pd.read_pickle(f"{DATA_PATH}emotion_audio_text.pkl")
df
```

```python
from typing import Union
def vid_path(split : str , name : str) -> Union[str,None]:
    video_path = f"{DATA_PATH}{split}_video/{name}.mp4"
    save_path = f"../../data/{split}_video/{name}.mp4"
    # print(video_path)
    if os.path.exists(video_path):
        return save_path
    else:
        return None
```

```python
from tqdm import tqdm

tqdm.pandas()

df['video_path'] = df.progress_apply(lambda x: vid_path(x[5], x[7]), axis=1)
df
```

```python
df.rename(columns={"path": "audio_path"} , inplace=True)
```
